```python
train = pd.read_csv("train_preprocessed", index_col = 0)
test = pd.read_csv("test_preprocessed", index_col = 0)
```

```python
train.dropna(inplace = True)
test.dropna(inplace = True)
```

```python
train_tweets = train["preprocessed tweets"]
test_tweets = test['preprocessed tweets']
```

```python
train_tweets = train["preprocessed tweets"].to_list()
test_tweets = test['preprocessed tweets'].to_list()
```

```python
train_labels = train["Sentiment"]
test_labels = test["Sentiment"]
```

```python
#Run this cell if lable encoder is not fit
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(train_labels)
y_train = le.transform(train_labels)
y_test = le.transform(test_labels)
class_label_mapping = dict(set(zip(y_train, train_labels.values)))
print(class_label_mapping)
dump(le, open('le.pkl', 'wb'))
```

```python
le = load(open('le.pkl', 'rb'))
y_train = le.transform(train_labels)
y_test = le.transform(test_labels)
class_label_mapping = dict(set(zip(y_train, train_labels.values)))
class_label_mapping
```

```python
num_words = None
oov_token = '<UNK>'
pad_type = 'post'
trunc_type = 'post'
# Tokenize our training data
tokenizer = Tokenizer(num_words=num_words, oov_token=oov_token)
tokenizer.fit_on_texts(train_tweets)

# Get our training data word index
word_index = tokenizer.word_index

# Encode training data sentences into sequences
train_sequences = tokenizer.texts_to_sequences(train_tweets)

# Get max training sequence length
maxlen = max([len(x) for x in train_sequences])

# Pad the training sequences
train_padded = pad_sequences(train_sequences, padding=pad_type, truncating=trunc_type, maxlen=maxlen)
```

```python
train_padded.shape
```

```python
# Encode testing data sentences into sequences
test_sequences = tokenizer.texts_to_sequences(test_tweets)

# Pad the testing sequences
test_padded = pad_sequences(test_sequences, padding=pad_type, truncating=trunc_type, maxlen=maxlen)
```

```python
test_padded.shape
```

```python
vocab_size = len(word_index) +1
```

```python
#getting glove embeddings for tokens
embeddings_index = dict()
f = open('glove.6B.300d.txt', encoding="utf8")
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()

embedding_matrix_glove = np.zeros((vocab_size, 300))
for word, i in word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix_glove[i] = embedding_vector
```

```python
#one hot encoding target
y_train_enc = to_categorical(y_train)
y_test_enc = to_categorical(y_test)
metric = F1Score(num_classes = 5, average = "weighted")
```

```python
emb_dim = 300 

#getting sentence 1 representations
input_ = Input(shape=(maxlen,)) #for sentence1 embeddings
emb = Embedding(input_dim = vocab_size, output_dim = emb_dim, input_length = maxlen, 
                 weights = [embedding_matrix_glove], trainable = False)(input_)
lstm1 = Bidirectional(LSTM(128, return_sequences = True, kernel_regularizer = l2(0.01)))(emb)
lstm2 = Bidirectional(LSTM(64, return_sequences = True, kernel_regularizer = l2(0.001)))(lstm1)
lstm3 = Bidirectional(LSTM(32))(lstm2)
# rs1 = tf.keras.layers.Reshape(target_shape=(maxlen,128,1))(lstm1)
# avg1 = tf.keras.layers.TimeDistributed(tf.keras.layers.GlobalAvgPool1D())(rs1)
# rs2 = tf.keras.layers.Reshape(target_shape=(maxlen,))(avg1)
dense1 = Dense(units = 128, activation = 'relu', kernel_regularizer = l2(0.001))(lstm3)
norm = BatchNormalization()(dense1)
dense2 = Dense(units = 64, activation = 'relu', kernel_regularizer = l2(0.001))(norm)
drop2 = Dropout(0.4)(dense2)
dense3 = Dense(units = 32, activation = 'relu', kernel_regularizer = l2(0.001))(drop2)
output = Dense(units = 5, activation = 'softmax')(dense3)
model0 = Model(inputs = input_, outputs = output)

```

```python
plot_model(model0, to_file='model0_sequence.png', show_shapes=True)
```

```python
model0.summary()
```

```python
from tensorflow.keras.callbacks import LearningRateScheduler, EarlyStopping, ModelCheckpoint
import tensorflow as tf
# learning rate schedule
def step_decay(epoch):

    import math
    initial_lrate = 0.001
    drop = 0.9
    epochs_drop = 10.0
    lrate = initial_lrate * math.pow(drop, math.floor((1+epoch)/epochs_drop))
    return lrate

lrate = LearningRateScheduler(step_decay)

es = EarlyStopping(monitor='val_accuracy', mode='max', patience = 3)
# fname = 'best_model3.hdf5'
# checkpoint = ModelCheckpoint(fname, monitor="val_accuracy", mode="max", save_best_only=True)
callbacks = [es, lrate]
```

```python
model0.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = metric)
```

```python
model0.fit(train_padded, y_train_enc, epochs = 50, batch_size = 128, 
           validation_data = (test_padded, y_test_enc), callbacks = callbacks)
```

```python
def get_classification_report(y_test, y_pred):
    from sklearn import metrics
    report = metrics.classification_report(y_test, y_pred, output_dict=True)
    df_classification_report = pd.DataFrame(report).transpose()
    #df_classification_report = df_classification_report.sort_values(by=['f1-score'], ascending=False)
    return df_classification_report
```

```python
pred = model0.predict(test_padded)
label_pred = np.array(tf.argmax(pred, axis = 1))
print("Performance on testing data:")
get_classification_report(y_test, label_pred)
```

```python
train_pred = model0.predict(train_padded)
train_label_pred = np.array(tf.argmax(train_pred, axis = 1))
print("Performance on training data:")
get_classification_report(y_train, train_label_pred)
```

# Custom Sel-Attention with Glove embeddings

```python
class Attention_Custom(tf.keras.layers.Layer):
    
    def __init__(self,att_units, inp_shape):
        super().__init__()
        self.att_units = att_units 
        self.inp_shape = inp_shape
        self.num_dim = self.inp_shape[-1] ##dimensions per time step
        self.ts = self.inp_shape[-2]  ##number of timesteps
        self.w = self.add_weight(shape = (self.num_dim,self.att_units), initializer='normal', trainable = True)
        self.b = self.add_weight(shape = (self.ts,self.att_units), initializer='zero', trainable = True)
        self.v = self.add_weight(shape = (self.att_units, self.num_dim), initializer = 'normal', trainable = True)
        self.v_b = self.add_weight(shape = (self.ts, self.num_dim))

    def call(self,x): 
        score = K.dot((K.tanh(K.dot(x,self.w)+self.b)),self.v)+self.v_b
        weights = K.softmax(score, axis=1)
        weighted_states = x*weights
        output = K.sum(weighted_states, axis=1)
        
        return weights, output
    
```

```python
emb_dim = 300 

#getting sentence 1 representations
input_ = Input(shape=(maxlen,))
emb = Embedding(input_dim = vocab_size, output_dim = emb_dim, input_length = maxlen, 
                 weights = [embedding_matrix_glove], trainable = False)(input_)
#lstm_s1 = Bidirectional(LSTM(256, return_sequences = True) , merge_mode = 'concat')(emb1)
lstm1 = Bidirectional(LSTM(256, return_sequences = True, 
                           dropout=0.3, kernel_regularizer=l2(1e-4)), merge_mode = 'concat')(emb)
lstm2 = Bidirectional(LSTM(128, return_sequences = True, 
                           dropout=0.3, kernel_regularizer=l2(1e-4)), merge_mode = 'concat')(lstm1)
att = Attention_Custom(128, lstm2.shape)
att_weights, context = att.call(lstm2)
dense1 = Dense(units = 128, activation = 'relu', kernel_regularizer = l2(1e-4))(context)
dense2 = Dense(units = 64, activation = 'relu', kernel_regularizer = l2(1e-4))(dense1)
dense3 = Dense(units = 32, activation = 'relu')(dense2)
output = Dense(units = 5, activation = 'softmax')(dense3)
model0 = Model(inputs = input_, outputs = output)
```
